### Vorbereitung :

Das hochgestellte Skript muss im selben Ordner sein wie das von euch zu bearbeitende XML File.

Ändert das zu parsende XML file im Befehl 
tree = ET.parse("LLL.xml" -- > hier sollte euer XML file name stehen. )

Zudem kann ich nicht garantieren, dass mein Skript genau so bei euch zu den richtigen Ergebnissen führt, evtl müsst ihr noch etwas umändern. Diese Änderungen findet ihr weiter unten. Ich gehe grundsätzlich von einem TEI- Codierten XML-File aus. 

                            
#### Informationen 

über alle vorkommenden Charaktere werden in in tags gegeben:
der Name in "name" tags, gender in "sex" tags.
der Text befindet sich in sp tags, wobei jeder sp tag einen speaker tag enthält, der 
Name der Sprechenden Person ist dann in einem "w" tag forhanden.
    
    <sp xml:id="sp-0202" who="#Portia_MV">
                        <speaker xml:id="spk-0202">
                            <w xml:id="w0032550">PORTIA</w>
                        </speaker>
                        
Bei meinem XML - File gab es einen Auslesefehler, wodurch vor jedes Element noch die TEI-NS Url geschrieben wurde. 
Wenn dieser Fehler nicht auftritt, müsst ihr ein paar Strings ändern, wird weiter unten noch beschrieben. 

Jetzt nachfolgend die Vorbereitung und Definition von Funktionen

```python
from xml.etree import ElementTree as ET
from collections import Counter

tree = ET.parse("LLL.xml")
root = tree.getroot()
personlist = root.findall('.//{http://www.tei-c.org/ns/1.0}person')
personlist = [x for x in personlist if x.find("{http://www.tei-c.org/ns/1.0}persName")]



def generatorparse(personlist):
    """Generatorfunktion über die iteriert werden kann;
    nimmt eine Liste aus Elementen entgegen und returned ein Tuple aus (name, sex)"""
    for x in personlist:
        name = x.find(".//{http://www.tei-c.org/ns/1.0}name").text.lower()
        sex = x.find(".//{http://www.tei-c.org/ns/1.0}sex").text.lower()
        yield (name, sex)
        
def speaktimes (root):
    """Erhält die root und returned dann ein Counter-dict, mit name : sprechzahlen Werten. 
    """
    allspeakers = [x.text.lower() for x in root.findall(".//{http://www.tei-c.org/ns/1.0}speaker/{http://www.tei-c.org/ns/1.0}w")]
    speakers = Counter(allspeakers)
    return speakers


def combine(speakcount, persons):
    """Kombiniert das Counter-dict und die Tuple aus generatorparse()
    returned eine Liste final, die Tuple mit (name, sex, speakcount) enthält. Wenn der Name nicht richtig gefunden wurde, wird 
    'speakcount not found' im Tuple gespeichert und muss dann von Hand eingefügt werden."""
    final = []
    for person in persons:
        name, sex = person
        if name in speakcount:
            final.append((name, sex, speakcount[name]))
        else: 
            final.append((name, sex, "speakcount not found!"))
    return final
            
            
print(personlist)
```

Das "print(personlist)" ist ein check, um zu sehen, ob Elemente gefunden werden.

Sollte bei obigem Teil eine leere Liste angezeigt werden, müsst man evtl die Suchtags ändern, mit dem nächsten Befehl werden alle Tagnamen ausgegeben. Somit sieht man, wo evtl Fehler sind. Passt die Suchtag-Strings in den find() Befehlen an, dann sollte es klappen.

Der untenstehende Befehl returned alle Tags, so könnt ihr sehen, wie eure tags benannt sind für den Parser. Checkt das ab und ändert eure Strings, wenn es nicht funktioniert.

Hier jetzt die wirkliche Ausgabe:

Die Ausgabe jetzt ist euer Ergebnis. Wenn Namen eurer Charaktere aus mehr als einem Wort bestehen oder ab und zu unterschiedlich benannt werden, müsst ihr im Excel-File diese Werte manuell noch eintragen, da das Skript diese nicht findet/beachtet. 

```python
persons = [ x for x in generatorparse(personlist) if x]
speakcount = speaktimes(root)
final = combine(speakcount, persons)
print(persons)
print(speakcount)
print(final)
```

Und jetzt werden die Daten in ein Excel file geschrieben. Es werden Spalten für antagonist etc erstellt, aber keine Werte eingetragen, dies muss man dann per Hand machen. 

```python
import xlsxwriter

# Create a workbook and add a worksheet.
workbook = xlsxwriter.Workbook('Theatredata.xlsx')
worksheet = workbook.add_worksheet()


# Start from the first cell. Rows and columns are zero indexed.
row = 1
col = 0
idnr = 0 

worksheet.write(0, 0,"ID")
worksheet.write(0, 1, "label")
worksheet.write(0, 2, "gender")
worksheet.write(0, 3, "per_ms_sps")
worksheet.write(0, 4, "role")
worksheet.write(0, 5, "importance")

# Iterate over the data and write it out row by row.
for name, sex, speakcount in final:
    worksheet.write(row, col, idnr)
    worksheet.write(row, col + 1, name)
    worksheet.write(row, col + 2, sex)
    worksheet.write(row, col + 3, speakcount)
    idnr += 1
    row += 1
workbook.close()
```
